```python
# Make pairwise correlation and cosine similarity matricies

from sklearn.metrics.pairwise import cosine_similarity

# Assume recom_df is already loaded and has the proper dtypes

# 1. Create an item-user pivot table:
#    Rows are customers, columns are items, and values are total quantity purchased.
item_user_matrix = recom_df.pivot_table(
    index='customer_id', 
    columns='product_id', 
    values='quantity', 
    aggfunc='sum', 
    fill_value=0
)

# 2. Compute the pairwise Pearson correlation matrix between items.
#    This will give you a DataFrame where each cell [i, j] represents the correlation between item i and item j.
pairwise_corr = item_user_matrix.corr(method='pearson')
print("Pairwise Pearson Correlation Matrix:")
print(pairwise_corr.head())

# 3. Compute the cosine similarity matrix between items.
#    Since the item vectors are in the columns, we need to take the transpose.
cosine_sim = cosine_similarity(item_user_matrix.T)
# Convert the result into a DataFrame for easier interpretation.
cosine_sim_matrix = pd.DataFrame(
    cosine_sim, 
    index=item_user_matrix.columns, 
    columns=item_user_matrix.columns
)
print("Cosine Similarity Matrix:")
print(cosine_sim_matrix.head())
```

```python
# Group by both customer_id and product_id, then aggregate:
# - ds_last_purchased: maximum date for each group
# - frequency: total quantity purchased for each group
customer_activities = (
    recom_df.groupby(['customer_id', 'product_id'])
            .agg(ds_last_purchased=('date', 'max'),
                 frequency=('quantity', 'sum'))
            .reset_index()
)

# Sort by customer_id ascending and frequency descending
customer_activities = customer_activities.sort_values(
    by=['customer_id', 'frequency'], ascending=[True, False]
)

# Display the result
print(customer_activities.head())
```

```python
# Get most recent price for each product_id
# Group by product_id and price, then take the max date as the last time that price was seen.
products = (
    recom_df.groupby(['product_id', 'price'])
              .agg(last_date=('date', 'max'))
              .reset_index()
)

# Optionally, you can rename the column for clarity:
products.rename(columns={'last_date': 'ds_last_price_update'}, inplace=True)

# Display the resulting table
print(products.head())
```

```python
# Get Pairwise receommendations 

# --- Assume these DataFrames already exist ---
# recom_df: the original transactions DataFrame
# products: DataFrame with unique product_id, price, ds_last_price_update
# customer_activities: aggregated customer activities table, e.g.:
# customer_activities = (
#     recom_df.groupby(['customer_id', 'product_id'])
#             .agg(ds_last_purchased=('date', 'max'),
#                  frequency=('quantity', 'sum'))
#             .reset_index()
# )
#
# pairwise_corr: item–item correlation DataFrame computed from a pivot table
# For example:
# item_user_matrix = recom_df.pivot_table(
#     index='customer_id', 
#     columns='product_id', 
#     values='quantity', 
#     aggfunc='sum', 
#     fill_value=0
# )
# pairwise_corr = item_user_matrix.corr(method='pearson')

# --- Step 1: Define a function to get the recommended product ---
def get_rec(product, corr_matrix):
    """
    Given a product id and a correlation matrix DataFrame (indexed and columned by product_id),
    return the recommended product: the one with the highest correlation with the input product,
    ignoring the product itself.
    """
    # Get the correlation series for the given product
    corr_series = corr_matrix.loc[product].copy()
    # Exclude self by setting the product's own correlation to -infinity
    corr_series[product] = -np.inf
    # The recommended product is the one with the highest correlation value
    recommended = corr_series.idxmax()
    return recommended

# --- Step 2: Create the cust_rec_pairwise table using customer_activities ---
# Start with the customer and unique product pair the customer bought.
cust_rec_pairwise = customer_activities[['customer_id', 'product_id']].copy()
cust_rec_pairwise = cust_rec_pairwise.drop_duplicates(subset=['customer_id', 'product_id'])

# For each purchased product, compute the recommended product using the correlation matrix.
cust_rec_pairwise['product_rec'] = cust_rec_pairwise['product_id'].apply(lambda p: get_rec(p, pairwise_corr))

# --- Step 3: Join with the products table to get the recommended product's price ---
# We want the price from the most recent price update. In the products table,
# if there are multiple rows per product (i.e. multiple prices over time),
# we select the row with the maximum ds_last_price_update per product_id.

# Compute the latest price for each product.
latest_products = products.loc[products.groupby('product_id')['ds_last_price_update'].idxmax()]

# For the join, keep only product_id and price, and rename price to product_rec_price.
latest_products = latest_products[['product_id', 'price']].rename(columns={'price': 'product_rec_price'})

# Merge on the recommended product_id (from cust_rec_pairwise) with the latest product info.
cust_rec_pairwise = cust_rec_pairwise.merge(
    latest_products,
    left_on='product_rec',
    right_on='product_id',
    how='left'
)

# Remove the extra product_id column that came from the merge and rename columns accordingly.
cust_rec_pairwise.drop(columns=['product_id_y'], inplace=True)
cust_rec_pairwise.rename(columns={'product_id_x': 'product_id'}, inplace=True)

# Optionally, reorder the columns as desired.
cust_rec_pairwise = cust_rec_pairwise[['customer_id', 'product_id', 'product_rec', 'product_rec_price']]

# Display the final table
print(cust_rec_pairwise.head())
```

```python
# Get cosine similarity recommedations 

# --- Assume these DataFrames already exist ---
# recom_df: original transactions DataFrame.
# products: DataFrame with columns: product_id, price, ds_last_price_update.
# customer_activities: DataFrame aggregated as follows:
#   customer_activities = (
#       recom_df.groupby(['customer_id', 'product_id'])
#               .agg(ds_last_purchased=('date', 'max'),
#                    frequency=('quantity', 'sum'))
#               .reset_index()
#   )
# cosine_sim_matrix: Square DataFrame with cosine similarity scores for each product pair,
#   with product_id as both the index and the columns.

# --- Step 1: Define a function to get the recommended product using cosine similarity ---
def get_rec(product, sim_matrix):
    """
    Given a product id and a similarity matrix (indexed and columned by product_id),
    return the recommended product: the one with the highest similarity score,
    ignoring the product itself.
    """
    # Get the similarity series for the given product.
    sim_series = sim_matrix.loc[product].copy()
    # Exclude the product itself by setting its own similarity to -infinity.
    sim_series[product] = -np.inf
    # Return the product id with the highest similarity.
    recommended = sim_series.idxmax()
    return recommended

# --- Step 2: Create the cust_rec_pairwise table using customer_activities ---
# Start with the unique customer-product pairs.
cust_rec_sim = customer_activities[['customer_id', 'product_id']].drop_duplicates()

# For each purchased product, compute the recommended product using the cosine similarity matrix.
cust_rec_sim['product_rec'] = cust_rec_sim['product_id'].apply(lambda p: get_rec(p, cosine_sim_matrix))

# --- Step 3: Prepare the latest price info from the products table ---
# For each product_id, select the row with the maximum ds_last_price_update
latest_products = products.loc[products.groupby('product_id')['ds_last_price_update'].idxmax()].copy()

# Keep only product_id and price, and rename price to product_rec_price for clarity.
latest_products = latest_products[['product_id', 'price']].rename(columns={'price': 'product_rec_price'})

# --- Step 4: Merge to obtain the recommended product's latest price ---
# Merge on the recommended product_id from cust_rec_pairwise with latest_products.
cust_rec_sim = cust_rec_sim.merge(
    latest_products,
    left_on='product_rec',
    right_on='product_id',
    how='left'
)

# Clean up the extra product_id column from the merge.
cust_rec_sim.drop(columns=['product_id_y'], inplace=True)
cust_rec_sim.rename(columns={'product_id_x': 'product_id'}, inplace=True)

# Optionally, reorder the columns.
cust_rec_sim = cust_rec_sim[['customer_id', 'product_id', 'product_rec', 'product_rec_price']]

# Display the final table.
print(cust_rec_sim.head())
```

```python
# Generate NMF pivot table

from sklearn.decomposition import NMF

# Step 1: Create the pivot table (customer-product matrix)
customer_product_matrix = recom_df.pivot_table(
    index='customer_id', 
    columns='product_id', 
    values='quantity', 
    aggfunc='sum', 
    fill_value=0
)

# Display the shape and a snippet of the pivot table to confirm its structure.
print("Customer-Product Matrix Shape:", customer_product_matrix.shape)
print(customer_product_matrix.head())

# Step 2: Apply Non-Negative Matrix Factorization (NMF)
# Choose the number of latent factors (components). Adjust n_components as needed.
n_components = 2
nmf_model = NMF(n_components=n_components, init='random', random_state=42)

# Fit the model and obtain the latent feature matrices:
#   W: shape (n_customers, n_components)
#   H: shape (n_components, n_products)
W = nmf_model.fit_transform(customer_product_matrix)
H = nmf_model.components_

# Optionally, view the dimensions of the factorized matrices.
print("W shape (customer factors):", W.shape)
print("H shape (product factors):", H.shape)

# (Optional) Reconstruct the approximate original matrix:
reconstructed_matrix = pd.DataFrame(
    data=W.dot(H),
    index=customer_product_matrix.index,
    columns=customer_product_matrix.columns
)

# Display a snippet of the reconstructed matrix.
print("Reconstructed Matrix (first 5 rows):")
print(reconstructed_matrix.head())
```

```python
# use NMF table to make recommendations 

# Start with unique (customer_id, product_id) pairs from customer_activities.
cust_rec_nmf = customer_activities[['customer_id', 'product_id']].drop_duplicates().copy()

# Define a function to get the recommended product based on the customer's row in reconstructed_matrix.
def get_rec_nmf(row):
    customer = row['customer_id']
    purchased_product = row['product_id']
    
    # Get the predicted scores for all products for this customer.
    scores = reconstructed_matrix.loc[customer].copy()
    
    # Exclude the product the customer already purchased.
    scores[purchased_product] = -np.inf
    
    # The recommended product is the one with the highest predicted score.
    recommended = scores.idxmax()
    return recommended

# Apply the recommendation function row-wise.
cust_rec_nmf['product_rec'] = cust_rec_nmf.apply(get_rec_nmf, axis=1)

# --- Step 4: Get the latest product price info ---
# If you have multiple price entries per product, select the most recent one.
latest_products = products.loc[
    products.groupby('product_id')['ds_last_price_update'].idxmax()
].copy()

# Keep only product_id and price, and rename price for clarity.
latest_products = latest_products[['product_id', 'price']].rename(columns={'price': 'product_rec_price'})

# --- Step 5: Merge to add the recommended product's price ---
cust_rec_nmf = cust_rec_nmf.merge(
    latest_products,
    left_on='product_rec',
    right_on='product_id',
    how='left'
)

# Remove the extra product_id column from the merge and rename columns appropriately.
cust_rec_nmf.drop(columns=['product_id_y'], inplace=True)
cust_rec_nmf.rename(columns={'product_id_x': 'product_id'}, inplace=True)

# Optionally, reorder the columns.
cust_rec_nmf = cust_rec_nmf[['customer_id', 'product_id', 'product_rec', 'product_rec_price']]

# Display the final recommendation table.
print(cust_rec_nmf.head())
```

```python
# Use embeddings to make deep learning model

import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Embedding, Flatten, Concatenate, Dense, Dropout, BatchNormalization
from tensorflow.keras.regularizers import l2
import numpy as np

# -----------------------------
# Step 1: Prepare the Training Data
# -----------------------------
# (Assume customer_product_matrix is already created.)
df_long = customer_product_matrix.stack().reset_index()
df_long.columns = ['customer_id', 'product_id', 'quantity']

# Create integer mappings for customers and products.
customer_ids = df_long['customer_id'].unique()
product_ids = df_long['product_id'].unique()
num_customers = len(customer_ids)
num_products = len(product_ids)

customer2idx = {cid: idx for idx, cid in enumerate(customer_ids)}
product2idx = {pid: idx for idx, pid in enumerate(product_ids)}

df_long['customer_idx'] = df_long['customer_id'].map(customer2idx)
df_long['product_idx'] = df_long['product_id'].map(product2idx)

# Prepare the training arrays.
X_customer = df_long['customer_idx'].values
X_product = df_long['product_idx'].values
y = df_long['quantity'].values.astype(np.float32)

# -----------------------------
# Step 2: Build the Neural Collaborative Filtering Model (Adjusted)
# -----------------------------
embedding_dim = 50   # Increased from 5 to 25 for richer embeddings.

# Customer embedding branch.
customer_input = Input(shape=(1,), name='customer_input')
customer_embedding = Embedding(input_dim=num_customers,
                               output_dim=embedding_dim,
                               embeddings_regularizer=l2(1e-4),
                               name='customer_embedding')(customer_input)
customer_vec = Flatten(name='flatten_customers')(customer_embedding)

# Product embedding branch.
product_input = Input(shape=(1,), name='product_input')
product_embedding = Embedding(input_dim=num_products,
                              output_dim=embedding_dim,
                              embeddings_regularizer=l2(1e-4),
                              name='product_embedding')(product_input)
product_vec = Flatten(name='flatten_products')(product_embedding)

# Concatenate embeddings.
concat = Concatenate(name='concatenate')([customer_vec, product_vec])

# Add Dense layers with regularization, batch normalization, and (reduced) dropout.
dense1 = Dense(128, activation='relu', kernel_regularizer=l2(1e-4), name='dense1')(concat)
bn1 = BatchNormalization(name='bn1')(dense1)
drop1 = Dropout(0.1, name='drop1')(bn1)  # Reduced dropout from 0.5 to 0.2

dense2 = Dense(64, activation='relu', kernel_regularizer=l2(1e-4), name='dense2')(drop1)
bn2 = BatchNormalization(name='bn2')(dense2)
drop2 = Dropout(0.1, name='drop2')(bn2)  # Reduced dropout from 0.3 to 0.2

output = Dense(1, name='output')(drop2)

# Define and compile the model.
model = Model(inputs=[customer_input, product_input], outputs=output)
optimizer = tf.keras.optimizers.Adam(learning_rate=0.0005)
model.compile(optimizer=optimizer, loss='mse')

# Print model summary.
model.summary()

# -----------------------------
# Step 3: Check GPU Availability and Train the Model
# -----------------------------
print("Num GPUs Available: ", len(tf.config.list_physical_devices('GPU')))

# Optionally, add an EarlyStopping callback.
early_stop = tf.keras.callbacks.EarlyStopping(monitor='val_loss', patience=3, restore_best_weights=True)

# Train the model.
history = model.fit(
    x=[X_customer, X_product],
    y=y,
    batch_size=256,
    epochs=30,  # Increase epochs to give the model more time.
    validation_split=0.2,
    callbacks=[early_stop]
)
```

```python
# Use deep learning model to make recommendations

# === Step 1: Create the deep prediction matrix ===
# Assume num_customers and num_products are defined, and customer_ids and product_ids
# are arrays of unique customer and product IDs corresponding to the indices.
# We create grid arrays to predict for every customer-product pair.

all_customers = np.arange(num_customers)    # customer indices (0, 1, 2, ..., num_customers-1)
all_products = np.arange(num_products)        # product indices (0, 1, 2, ..., num_products-1)

# Create a grid of customer and product indices.
customer_grid = np.repeat(all_customers, num_products)
product_grid = np.tile(all_products, num_customers)

# Predict scores for all customer-product pairs.
preds = model.predict([customer_grid, product_grid], batch_size=256)

# Reshape the predictions to form a matrix with shape (num_customers, num_products)
deep_pred_matrix_np = preds.reshape(num_customers, num_products)

# Create a DataFrame for easier lookup.
# Use customer_ids as the index and product_ids as the columns.
deep_pred_matrix = pd.DataFrame(deep_pred_matrix_np, index=customer_ids, columns=product_ids)

# === Step 2: Generate the deep recommendation table (cust_rec_deep) ===
# Start with unique (customer_id, product_id) pairs from customer_activities.
cust_rec_deep = customer_activities[['customer_id', 'product_id']].drop_duplicates().copy()

# Define a function to get the recommended product using deep_pred_matrix.
def get_rec_deep(row):
    customer = row['customer_id']
    purchased_product = row['product_id']
    
    # Get the prediction scores for all products for this customer.
    scores = deep_pred_matrix.loc[customer].copy()
    
    # Exclude the purchased product.
    scores[purchased_product] = -np.inf
    
    # The recommended product is the one with the highest predicted score.
    recommended = scores.idxmax()
    return recommended

# Apply the recommendation function row-wise.
cust_rec_deep['product_rec'] = cust_rec_deep.apply(get_rec_deep, axis=1)

# === Step 3: Merge with the latest product price information ===
# Assume 'products' DataFrame has columns: product_id, price, ds_last_price_update.
# Select the most recent price per product.
latest_products = products.loc[
    products.groupby('product_id')['ds_last_price_update'].idxmax()
].copy()

# Keep only product_id and price, and rename price for clarity.
latest_products = latest_products[['product_id', 'price']].rename(columns={'price': 'product_rec_price'})

# Merge on the recommended product.
cust_rec_deep = cust_rec_deep.merge(
    latest_products,
    left_on='product_rec',
    right_on='product_id',
    how='left'
)

# Remove the extra product_id column from the merge and clean up column names.
cust_rec_deep.drop(columns=['product_id_y'], inplace=True)
cust_rec_deep.rename(columns={'product_id_x': 'product_id'}, inplace=True)

# Optionally, reorder the columns.
cust_rec_deep = cust_rec_deep[['customer_id', 'product_id', 'product_rec', 'product_rec_price']]

# Display the final recommendation table.
print(cust_rec_deep.head())
```
